Why are some possible reasons as to why a person's employment title may have not been provided?

```python
#Create some examples that represent the case that we want to clean up
examples = ['owner', 'Supervisor', ' Project Manage', np.nan]  # Supervisor is correct but the rest need to be corrected 
```

```python
# Write a function to clean up these use cases and increase uniformity.
def clean_title(job_title):
  if isinstance(job_title, str): #isinstance is a built in Python funtion to determine whether or not a given value belongs to a cetrain data type
    return job_title.title().strip() #.title (built in Python method) will capitalize the first letter.
  else:
    return "Unknown"
```

```python
#create a for loop
for title in examples:
  print(clean_title(title))
```

```python
# list comprehensions can combine function calls and for loops over lists
# into one succinct and fairly readable single line of code.
[clean_title(title) for title in examples]
```

```python
# We have a function that works as expected. Lets apply it to our column.
# This time we'll overwrite the original column

df1['emp_title'] = df1['emp_title'].apply(clean_title)
df1.head()
```
